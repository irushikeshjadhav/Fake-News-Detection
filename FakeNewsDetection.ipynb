```python
import pandas as pd
import numpy as np

true=pd.read_csv('True.csv')
fake=pd.read_csv('Fake.csv')
```

```python
true.head()
```

```python
fake.head()
```

```python
true['label']=1
fake['label']=0
```

```python
true.head()
```

```python
fake.head()
```

```python
news = pd.concat([fake,true],axis=0)
```

```python
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news=news.drop(['title','subject','date'],axis=1)
```

```python
news.head()
```

```python
news = news.sample(frac=1)  #Reshuffling
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'],axis=1,inplace=True)
```

```python
news.head()
```

```python
import re

def wordopt(text):
    text = text.lower() #convert text into lowercase

    #remove URLs
    text = re.sub(r'https?://\S+|www\.\S+','',text) #pattern for urls

    #remove Html tags
    text = re.sub(r'<.*?>','',text)

    #remove punctuation
    text = re.sub(r'[^\w\s]',' ',text)

    #remove digits
    text = re.sub(r'\d','',text)

    #remove newline characters
    text = re.sub(r'\n',' ',text)

    return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(xv_train,y_train)
```

```python
pred_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rfc))
```

```python
def output_label(n):
    if n == 0:
        return "It is a Fake News"
    elif n == 1:
        return "It is a Genuine News"
```

```python
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_RFC = rfc.predict(new_xv_test)

    return "\n\n LR Prediction : {} \n DT Prediction : {} \n RFC Prediction : {}".format(output_label(pred_LR[0]),
                                                                                       output_label(pred_DT[0]),
                                                                                       output_label(pred_RFC[0]))
```

```python
news_article = str(input())
manual_testing(news_article)
```
